```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import os
%pylab inline
figsize(12, 8)
```

```python
def massage_data(run):
    # Change time into time values
    run.linuxtime = run.linuxtime.apply(datetime.datetime.fromtimestamp)
    
    
    run.set_index(run.linuxtime, drop=True, inplace=True)
    del run["linuxtime"]
    for col in run:
        run[col].index = run.index
    
    # Interpolate what we can
    run.interpolate(method='time')
    
    # Fill forward for other values
    run.fillna(method='pad', inplace=True)
    # fill backward for initial values
    run.fillna(method='bfill', inplace=True)

    return run
```

```python
run = pd.read_csv("test_logs/2016-07-06_22_run0.csv")
run = run.append(pd.read_csv("test_logs/2016-07-06_22_run1.csv"))
run = massage_data(run)
run.ds_time = run.ds_time.apply(datetime.datetime.fromtimestamp)
```

```python
run.an_300v_cur.plot(markevery=10, style='y')
run.an_300v_cur.rolling(100).mean(center=True).plot(style='r')
```
