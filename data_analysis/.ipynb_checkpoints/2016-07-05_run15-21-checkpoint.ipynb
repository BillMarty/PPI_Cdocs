```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import datetime
%pylab inline
figsize(12, 8)
```

```python
def massage_data(run):
    # Filter some nonsense values
    run.linuxtime = run.linuxtime.apply(datetime.datetime.fromtimestamp)
    run.rpm = run.rpm.apply(lambda x: x if x < 10000 and x >= 0 else np.nan)
    run.ds_volt = run.ds_volt.apply(lambda x: x if x < 400 and x > 100 else np.nan)
    run.ds_cur_300v = run.ds_cur_300v.apply(lambda x: x if x < 1e8 else np.nan)
    
    run.set_index(run.linuxtime, drop=True, inplace=True)
    del run["linuxtime"]
    del run["time"]
    for col in run:
        run[col].index = run.index
    
    # Interpolate what we can
    run.interpolate(method='time')
    
    # Fill forward for other values
    run.fillna(method='pad', inplace=True)

    return run
```

```python
i = 15
runs = []
while os.path.exists("test_logs/2016-07-05_run%d.csv"%(i)):
    run = pd.read_csv("test_logs/2016-07-05_run%d.csv"%(i))

    run = massage_data(run)
    runs.append(run)
    i += 1
```

```python
flag = False
for r in runs:
    if flag:
        run = run.append(r)
    else:
        run = r
        flag = True
```

```python
def fit_and_plot(x, y):
    fit = np.polyfit(x, y, 1)
    plt.plot(x, y, '.')
    plt.plot(x, np.poly1d(fit)(x))
    return fit
```

```python
fit_and_plot(run.an_cur_300v[run.ds_cur_300v > 2], run.ds_cur_300v[run.ds_cur_300v > 2])
```
